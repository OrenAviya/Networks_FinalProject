# Calculation of the PDF function of the message delays

## example on the messages 

```python
path_name = "C:/Users/97258/OneDrive - Ariel University/הודעות טקסט 10 דק.xlsx"
```

```python
df = pd.read_excel(path_name)
```

```python
df
```

```python
column_name = 'מרחק בין הנוכחי לקודם בזמן'

# Filter out NaN values from the column
valid_values = df[column_name].dropna()

# Calculate the histogram using only valid values
counts, bin_edges = np.histogram(valid_values, density=True)
```

```python
print(counts)
print(bin_edges)
```

```python
pdf = counts / sum(counts)
```

```python
# Calculate the bin widths
bin_widths = bin_edges[1:] - bin_edges[:-1]

# Calculate the PDF values
pdf_values = counts * bin_widths

# Create a bar plot to visualize the PDF with pink bars
plt.bar(bin_edges[:-1], pdf_values, width=bin_widths, align='edge', color='pink', alpha=0.5, label='Histogram')
plt.xlabel('Time Delay')  # Set the X-axis label
plt.ylabel('Probability Density')

# Plot a line over the bins to visualize the PDF
plt.plot(bin_edges[:-1], pdf_values, marker='o', color='red', linestyle='dashed', linewidth=2, markersize=6, label='PDF Line')

plt.title('Probability Density Function')
plt.legend()
plt.show()
```
